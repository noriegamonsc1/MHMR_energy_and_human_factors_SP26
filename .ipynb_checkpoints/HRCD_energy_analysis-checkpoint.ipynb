# Dataset Description
## HRCD Energy Logging — Human-Robot Collaborative Disassembly

**Project:** Energy analysis for Life Cycle Analysis (LCA)  
**Robot:** Mobile manipulator (Clearpath Warthog base + dual UR arms + Robotiq grippers)  
**ROS version:** Noetic (Ubuntu 20.04)  
**Logger node:** `full_energy_logger.py`

---

## Files

| File | Description |
|---|---|
| `energy_log_left_<timestamp>.csv` | Left UR arm — joint states, power, wrench |
| `energy_log_right_<timestamp>.csv` | Right UR arm — joint states, power, wrench |
| `energy_log_base_<timestamp>.csv` | Mobile base — battery power and temperatures |

The `<timestamp>` suffix is a Unix epoch integer (seconds since January 1, 1970) shared across all three files from the same recording session, so files can be matched and merged on the `timestamp` column.

---

## Arm files (`energy_log_left_*.csv`, `energy_log_right_*.csv`)

Each row represents one sample from `/joint_states` for that arm, recorded at approximately **125 Hz**.

### Timing

| Column | Unit | Description |
|---|---|---|
| `timestamp` | s (Unix epoch) | Wall-clock time on the logging workstation at the moment the `/joint_states` message was received, obtained via `rospy.Time.now().to_sec()`. Used instead of `msg.header.stamp` to avoid clock-synchronisation artifacts between the robot PC and the workstation. |

### Joint positions (pos)

Six columns, one per UR arm joint in the order reported by the UR driver.

| Column | Unit | Joint |
|---|---|---|
| `pos_0` | rad | Shoulder pan |
| `pos_1` | rad | Shoulder lift |
| `pos_2` | rad | Elbow |
| `pos_3` | rad | Wrist 1 |
| `pos_4` | rad | Wrist 2 |
| `pos_5` | rad | Wrist 3 |

### Joint velocities

Six columns of **numerically differentiated** velocities. Because `/joint_states` on the UR driver reports zero or near-zero velocity even during motion, velocities are computed in the logger as:

$$\omega_i(t) =\frac{\Delta \text{pos}_i}{\Delta t}$$

where $\Delta t$ is the difference between consecutive wall-clock timestamps.

| Column | Unit | Description |
|---|---|---|
| `vel_0` | rad/s | Shoulder pan velocity |
| `vel_1` | rad/s | Shoulder lift velocity |
| `vel_2` | rad/s | Elbow velocity |
| `vel_3` | rad/s | Wrist 1 velocity |
| `vel_4` | rad/s | Wrist 2 velocity |
| `vel_5` | rad/s | Wrist 3 velocity |

> **Note:** Numerical differentiation amplifies high-frequency noise. Apply a rolling-mean smoothing window (e.g. 25 samples ≈ 0.2 s at 125 Hz) before using these values for power or cycle detection.

### Joint efforts

Six columns of joint torques as reported directly by the UR hardware interface. These are gravity-compensation torques when the arm is stationary, and the combined load torque during motion.

| Column | Unit | Description |
|---|---|---|
| `eff_0` | N·m | Shoulder pan effort |
| `eff_1` | N·m | Shoulder lift effort |
| `eff_2` | N·m | Elbow effort |
| `eff_3` | N·m | Wrist 1 effort |
| `eff_4` | N·m | Wrist 2 effort |
| `eff_5` | N·m | Wrist 3 effort |


### Power

| Column | Unit | Description |
|---|---|---|
| `arm_power` | W | Instantaneous mechanical power of the arm, computed as $P_{arm} = \sum_{i=0}^{5} \tau_i \cdot \omega_i$, where $\tau_i$ is `eff_i` and $\omega_i$ is `vel_i`. This is a proxy for electrical power — it does not account for motor efficiency losses or back-EMF. Negative values indicate regenerative braking (the arm is decelerating against a load). |
| `tool_power` | W | Instantaneous electrical power consumed by the tool (gripper) at the flange, computed as $P_{tool} = I_{tool} \times V_{tool}$, where both values come from `/ur_hardware_interface/tool_data`. This captures gripper holding current even when the arm is stationary. |

### Context channels

| Column | Unit | Description |
|---|---|---|
| `speed_scaling` | dimensionless (0–1) | UR speed scaling factor from `/ur/speed_scaling_factor`. A value of 1.0 means the arm is executing at full programmed speed. Values below 1.0 indicate protective slowdown (e.g. due to safety zone proximity or manual speed override). Always 1.0 in nominal operation. |
| `wrench_fx` | N | End-effector force in the X direction, from `/ur/wrench` (F/T sensor at the flange). |
| `wrench_fy` | N | End-effector force in the Y direction. |
| `wrench_fz` | N | End-effector force in the Z direction. At rest this approximates the weight of the gripper and payload (~9–10 N ≈ 1 kg). Changes during grasping indicate contact forces. |
| `wrench_tx` | N·m | End-effector torque about the X axis. |
| `wrench_ty` | N·m | End-effector torque about the Y axis. |
| `wrench_tz` | N·m | End-effector torque about the Z axis. |
| `in_cycle` | 0 or 1 | Cycle activity flag from `/cycle_active`. `1` = robot is executing a pick-and-drop cycle (the LCA functional unit). `0` = idle or transitioning. **In current recordings this column is always 0** because the `/cycle_active` publisher was not running. Cycles are instead inferred from the velocity signal in post-processing (see `HRCD_energy_analysis.ipynb`). |

---

## Base file (`energy_log_base_*.csv`)

Each row is written in sync with the left arm's `/joint_states` publish rate (~125 Hz). Battery and temperature values are the most recently received reading from their respective topics at the time of writing — they are **not interpolated**.

| Column | Unit | Source topic | Description |
|---|---|---|---|
| `timestamp` | s (Unix epoch) | `rospy.Time.now()` | Same wall-clock timestamp as the arm files. |
| `left_voltage` | V | `/left_drive/status/battery_voltage` | Terminal voltage of the left motor drive battery. |
| `left_current` | A | `/left_drive/status/battery_current` | Current drawn by the left motor drive. Positive = discharging. |
| `right_voltage` | V | `/right_drive/status/battery_voltage` | Terminal voltage of the right motor drive battery. |
| `right_current` | A | `/right_drive/status/battery_current` | Current drawn by the right motor drive. |
| `base_power` | W | Computed | Instantaneous power of the mobile base: $P_{base} = V_{left} \cdot I_{left} + V_{right} \cdot I_{right}$. Zero when the base is stationary and no battery topics are received. |
| `left_motor_temp` | °C (integer) | `/left_drive/status/motor_temperature` | Left drive motor winding temperature. Useful for detecting thermal losses under sustained load. |
| `right_motor_temp` | °C (integer) | `/right_drive/status/motor_temperature` | Right drive motor winding temperature. |
| `left_heatsink_temp` | °C (integer) | `/left_drive/status/heatsink_temperature` | Left motor controller heatsink temperature. |
| `right_heatsink_temp` | °C (integer) | `/right_drive/status/heatsink_temperature` | Right motor controller heatsink temperature. |
| `in_cycle` | 0 or 1 | `/cycle_active` | Same cycle flag as in the arm files. |

---

## Energy calculation

Energy is computed via the **trapezoidal rule** over any time window $[t_0, t_1]$:

$$E = \int_{t_0}^{t_1} P(t)\, dt \approx \sum_{k} \frac{P_k + P_{k+1}}{2} \cdot (t_{k+1} - t_k)$$

In Python:
```python
import numpy as np
E_joules = np.trapezoid(df['arm_power'].values, df['t_rel'].values)
```

To convert to watt-hours: $E_{Wh} = E_J / 3600$

---

## Known limitations

| Issue | Impact | Status |
|---|---|---|
| Velocity is numerically differentiated, not measured | `arm_power` is noisy at high frequency and underestimates power at low speeds | Mitigated by smoothing; will improve if UR velocity feedback is fixed upstream |
| `in_cycle` always 0 | Cycle boundaries must be inferred from velocity | Workaround in place in analysis notebook; fix requires publishing `/cycle_active` from task controller |
| `base_power` always 0 in arm files | Mobile base energy not captured when base is stationary | Expected behavior; base file should be used for base energy |
| Wall-clock timestamp | Small jitter (~1–2 ms) between arm left/right files | Acceptable for LCA; merge files with `pd.merge_asof` with `tolerance=0.01` |
| No BMS pack-level power | Total system energy cannot be cross-validated against ground truth | `valence_bms_msgs` not built on workstation; future work |

# HRCD Energy Analysis — Left Arm
Human-Robot Collaborative Disassembly · Life Cycle Analysis

- **Functional unit:** one pick-and-drop cycle  
- **Data:** `energy_log_left_<timestamp>.csv` (positions, numerically-differentiated velocities, efforts, tool power, wrench)  
**Note:** `in_cycle` flag was not published during this recording. This notebook detects cycles automatically from the velocity signal.

## 0 · Imports and configuration

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy.signal import find_peaks

plt.rcParams.update({
    'figure.dpi': 120,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.grid': True,
    'grid.alpha': 0.3,
    'font.size': 10,
})

# ── USER SETTINGS ────────────────────────────────────────────────────────────
DATA_FILE      = 'sample_50K_left.csv'   # path to your left-arm CSV
VEL_THRESHOLD  = 0.15   # rad/s — below this the arm is considered stationary
SMOOTH_WINDOW  = 25     # samples for rolling-mean velocity smoothing (~0.2 s at 125 Hz)
MIN_SEG_DUR    = 1.0    # s — motion segments shorter than this are ignored
CYCLE_GAP      = 10.0   # s — gaps larger than this separate distinct cycles
# ─────────────────────────────────────────────────────────────────────────────
````

## 1 · Load and preprocess

````python
df = pd.read_csv(DATA_FILE)
print(f'Rows: {len(df):,}   Columns: {len(df.columns)}')
print(f'Columns: {df.columns.tolist()}')
df.head(3)
````

````python
# ── Relative time ─────────────────────────────────────────────────────────────
t0 = df['timestamp'].min()
df['t_rel'] = df['timestamp'] - t0
duration_s = df['t_rel'].max()
print(f'Recording duration: {duration_s:.1f} s  ({duration_s/60:.2f} min)')

# ── Sample-rate diagnostics ───────────────────────────────────────────────────
dt = df['timestamp'].diff().dropna()
print(f'Mean Δt: {dt.mean()*1000:.2f} ms  →  {1/dt.mean():.1f} Hz')
print(f'Out-of-order timestamps: {(dt < 0).sum()}')
print(f'Gaps > 50 ms: {(dt > 0.05).sum()}')

# ── Velocity magnitude (sum of |vel_i| across all 6 joints) ──────────────────
vel_cols = [f'vel_{i}' for i in range(6)]
eff_cols = [f'eff_{i}' for i in range(6)]
pos_cols = [f'pos_{i}' for i in range(6)]

df['vel_mag'] = df[vel_cols].abs().sum(axis=1)
df['vel_smooth'] = df['vel_mag'].rolling(SMOOTH_WINDOW, center=True).mean().fillna(df['vel_mag'])
df['eff_mag'] = df[eff_cols].abs().sum(axis=1)
print(f'\nVelocity magnitude — mean: {df["vel_mag"].mean():.3f}  max: {df["vel_mag"].max():.3f} rad/s')
````

## 2 · Energy calculation

Power is estimated per joint as:

$$P_{arm}(t) = \sum_{i=1}^{6} \tau_i(t) \cdot \omega_i(t)$$

where $\tau_i$ is the joint effort (N·m) and $\omega_i$ is the numerically-differentiated joint velocity (rad/s).

Energy is then computed via the **trapezoidal rule**:

$$E = \int_{t_0}^{t_1} P(t)\, dt \approx \sum_{k} \frac{P_k + P_{k+1}}{2} \cdot \Delta t_k$$

Tool power (gripper holding current) is logged separately as $P_{tool} = I_{tool} \times V_{tool}$ and added to get total arm-side energy.

````python
def energy_J(power_series, time_series):
    """Trapezoidal integration of power over time → energy in Joules."""
    return np.trapezoid(power_series.values, time_series.values)

E_arm_total  = energy_J(df['arm_power'],  df['t_rel'])
E_tool_total = energy_J(df['tool_power'], df['t_rel'])
E_total      = E_arm_total + E_tool_total

print('=== Total energy over full recording ===')
print(f'  Arm (τ·ω):     {E_arm_total:>10.2f} J   ({E_arm_total/3600:.5f} Wh)')
print(f'  Tool (gripper):{E_tool_total:>10.2f} J   ({E_tool_total/3600:.5f} Wh)')
print(f'  Total:         {E_total:>10.2f} J   ({E_total/3600:.5f} Wh)')
print(f'\nMean arm power:  {df["arm_power"].mean():.3f} W')
print(f'Peak arm power:  {df["arm_power"].max():.3f} W')
print(f'Mean tool power: {df["tool_power"].mean():.3f} W')
````

## 3 · Overview plots

````python
fig, axes = plt.subplots(4, 1, figsize=(14, 10), sharex=True)

# 3a — Joint positions
for i in range(6):
    axes[0].plot(df['t_rel'], df[f'pos_{i}'], lw=0.5, alpha=0.8, label=f'J{i}')
axes[0].set_ylabel('Position (rad)')
axes[0].legend(ncol=6, fontsize=8, loc='upper right')
axes[0].set_title('Joint positions')

# 3b — Smoothed velocity magnitude
axes[1].plot(df['t_rel'], df['vel_smooth'], lw=0.7, color='#378ADD')
axes[1].axhline(VEL_THRESHOLD, color='#E24B4A', lw=1, ls='--', label=f'threshold {VEL_THRESHOLD} rad/s')
axes[1].set_ylabel('Vel magnitude (rad/s)')
axes[1].legend(fontsize=8)
axes[1].set_title('Smoothed velocity magnitude')

# 3c — Arm power
axes[2].plot(df['t_rel'], df['arm_power'], lw=0.5, color='#1D9E75', alpha=0.9)
axes[2].axhline(0, color='gray', lw=0.5)
axes[2].set_ylabel('Power (W)')
axes[2].set_title('Arm power (τ·ω)')

# 3d — Tool power
axes[3].plot(df['t_rel'], df['tool_power'], lw=0.5, color='#EF9F27')
axes[3].set_ylabel('Power (W)')
axes[3].set_xlabel('Time (s)')
axes[3].set_title('Tool power (gripper)')

plt.tight_layout()
plt.savefig('overview.png', bbox_inches='tight')
plt.show()
````

## 4 · Automatic cycle detection

Since `in_cycle` was not published, cycles are inferred from the velocity signal:

1. Smooth the velocity magnitude with a rolling mean.
2. Threshold: any row with `vel_smooth > VEL_THRESHOLD` is labelled **moving**.
3. Consecutive moving rows form a **motion segment**.
4. Motion segments separated by gaps ≤ `CYCLE_GAP` seconds belong to the **same cycle**.
5. Each cluster of segments = one pick-and-drop cycle.

````python
# ── Step 1-3: motion segments ─────────────────────────────────────────────────
df['moving'] = df['vel_smooth'] > VEL_THRESHOLD
df['seg_id'] = (df['moving'] != df['moving'].shift()).cumsum()

seg_rows = []
for seg_id, grp in df[df['moving']].groupby('seg_id'):
    dur = grp['t_rel'].iloc[-1] - grp['t_rel'].iloc[0]
    if dur < MIN_SEG_DUR:
        continue
    seg_rows.append({
        'seg_id':    seg_id,
        't_start':   grp['t_rel'].iloc[0],
        't_end':     grp['t_rel'].iloc[-1],
        'duration':  dur,
        'peak_vel':  grp['vel_smooth'].max(),
        'mean_power': grp['arm_power'].mean(),
        'arm_energy_J': np.trapezoid(grp['arm_power'].values, grp['t_rel'].values),
        'tool_energy_J': np.trapezoid(grp['tool_power'].values, grp['t_rel'].values),
    })

segs = pd.DataFrame(seg_rows)
print(f'Motion segments (duration > {MIN_SEG_DUR}s): {len(segs)}')
print(segs[['t_start','t_end','duration','peak_vel','arm_energy_J']].round(2).to_string())
````

````python
# ── Step 4-5: cluster segments into cycles ────────────────────────────────────
segs = segs.sort_values('t_start').reset_index(drop=True)
segs['gap_before'] = segs['t_start'].diff().fillna(999)
segs['cycle_id'] = (segs['gap_before'] > CYCLE_GAP).cumsum()

cycles = segs.groupby('cycle_id').agg(
    t_start      = ('t_start', 'min'),
    t_end        = ('t_end',   'max'),
    n_segments   = ('seg_id',  'count'),
    peak_vel     = ('peak_vel','max'),
    arm_energy_J = ('arm_energy_J', 'sum'),
    tool_energy_J= ('tool_energy_J','sum'),
).reset_index(drop=True)

cycles['duration']    = cycles['t_end'] - cycles['t_start']
cycles['total_energy_J'] = cycles['arm_energy_J'] + cycles['tool_energy_J']

# Include tool energy during rest (gripper holding between cycles)
# We allocate rest-period tool energy proportionally to each cycle
total_tool_J = energy_J(df['tool_power'], df['t_rel'])
motion_tool_J = cycles['tool_energy_J'].sum()
rest_tool_J = total_tool_J - motion_tool_J
cycles['tool_energy_rest_J'] = rest_tool_J / len(cycles)   # distributed evenly
cycles['total_energy_incl_rest_J'] = cycles['arm_energy_J'] + cycles['tool_energy_J'] + cycles['tool_energy_rest_J']

print(f'Detected cycles: {len(cycles)}')
print(f'\nCycle breakdown:')
print(cycles[['t_start','t_end','duration','n_segments','arm_energy_J',
              'tool_energy_J','tool_energy_rest_J','total_energy_incl_rest_J']].round(2).to_string())
````

````python
# ── Per-cycle summary stats ────────────────────────────────────────────────────
print('=== Per-cycle energy summary (motion only) ===')
print(f"  Mean arm energy/cycle:   {cycles['arm_energy_J'].mean():.2f} J")
print(f"  Std  arm energy/cycle:   {cycles['arm_energy_J'].std():.2f} J")
print(f"  Mean total energy/cycle: {cycles['total_energy_incl_rest_J'].mean():.2f} J")
print(f"  Std  total energy/cycle: {cycles['total_energy_incl_rest_J'].std():.2f} J")
print(f"  Mean cycle duration:     {cycles['duration'].mean():.2f} s")
print(f"  Std  cycle duration:     {cycles['duration'].std():.2f} s")
print(f"\nLCA functional unit — energy per pick-and-drop cycle:")
mean_E = cycles['total_energy_incl_rest_J'].mean()
print(f"  {mean_E:.2f} J   ({mean_E/3600*1000:.4f} mWh)")
````

## 7 · Joint-level contribution

````python
# Per-joint energy over full recording
joint_energies = {}
for i in range(6):
    p = df[f'eff_{i}'] * df[f'vel_{i}']
    joint_energies[f'J{i}'] = np.trapezoid(p.values, df['t_rel'].values)

je = pd.Series(joint_energies)
print('=== Joint energy contributions (full recording) ===')
print(je.round(2).to_string())
print(f'Sum: {je.sum():.2f} J  (matches arm_power integral: {E_arm_total:.2f} J)')

fig, ax = plt.subplots(figsize=(7, 4))
colors_bar = ['#378ADD' if v >= 0 else '#E24B4A' for v in je.values]
ax.bar(je.index, je.values, color=colors_bar, alpha=0.85)
ax.axhline(0, color='gray', lw=0.5)
ax.set_ylabel('Energy (J)')
ax.set_title('Energy contribution per joint (full recording)')
plt.tight_layout()
plt.savefig('joint_energy.png', bbox_inches='tight')
plt.show()
````

Zoom into one cycle to inspect the power profile in detail.

````python
CYCLE_IDX = 0   # ← change to inspect a different cycle (0-indexed)

cyc = cycles.iloc[CYCLE_IDX]
mask = (df['t_rel'] >= cyc['t_start'] - 1) & (df['t_rel'] <= cyc['t_end'] + 1)
d = df[mask].copy()

fig, axes = plt.subplots(3, 1, figsize=(11, 8), sharex=True)

# Velocity
axes[0].plot(d['t_rel'], d['vel_smooth'], color='#888780', lw=1)
axes[0].axhline(VEL_THRESHOLD, color='#E24B4A', lw=1, ls='--')
axes[0].axvspan(cyc['t_start'], cyc['t_end'], alpha=0.12, color='#378ADD')
axes[0].set_ylabel('Vel mag (rad/s)')
axes[0].set_title(f'Cycle {CYCLE_IDX+1} deep-dive  |  arm: {cyc["arm_energy_J"]:.2f} J  tool: {cyc["tool_energy_J"]:.2f} J')

# Per-joint velocity
for i in range(6):
    axes[1].plot(d['t_rel'], d[f'vel_{i}'], lw=0.7, alpha=0.8, label=f'J{i}')
axes[1].axhline(0, color='gray', lw=0.5)
axes[1].axvspan(cyc['t_start'], cyc['t_end'], alpha=0.12, color='#378ADD')
axes[1].set_ylabel('Joint vel (rad/s)')
axes[1].legend(ncol=6, fontsize=8)

# Arm power + cumulative energy
ax2b = axes[2].twinx()
axes[2].plot(d['t_rel'], d['arm_power'], color='#1D9E75', lw=0.8, label='arm power')
axes[2].plot(d['t_rel'], d['tool_power'], color='#EF9F27', lw=0.8, label='tool power')
axes[2].axhline(0, color='gray', lw=0.5)
axes[2].axvspan(cyc['t_start'], cyc['t_end'], alpha=0.12, color='#378ADD')

# Cumulative energy within cycle
dt_cyc = d['t_rel'].diff().fillna(0)
cum_E = (d['arm_power'] * dt_cyc).cumsum()
ax2b.plot(d['t_rel'], cum_E, color='#534AB7', lw=1, ls='--', label='cumulative E')
ax2b.set_ylabel('Cum. energy (J)', color='#534AB7')
ax2b.tick_params(axis='y', labelcolor='#534AB7')

axes[2].set_ylabel('Power (W)')
axes[2].set_xlabel('Time (s)')
lines1, labs1 = axes[2].get_legend_handles_labels()
lines2, labs2 = ax2b.get_legend_handles_labels()
axes[2].legend(lines1+lines2, labs1+labs2, fontsize=8)

plt.tight_layout()
plt.savefig(f'cycle_{CYCLE_IDX+1}_deepdive.png', bbox_inches='tight')
plt.show()
````

## 10 · Notes and caveats

1. **Velocity is numerically differentiated** from position. High-frequency noise is suppressed by the rolling-mean smoother but spikes in power may still occur at rapid direction changes. If noise is excessive, increase `SMOOTH_WINDOW`.

2. **`in_cycle` was not published** during this recording. Cycle boundaries are inferred from the velocity threshold and gap criterion (`CYCLE_GAP`). Verify visually that the detected spans match the actual pick-and-drop events. Once `/cycle_active` is publishing reliably, replace the cycle detection block with:
   ```python
   df['cycle_id'] = (df['in_cycle'].diff() == 1).cumsum()
   df.loc[df['in_cycle'] == 0, 'cycle_id'] = np.nan
   ```

3. **Tool (gripper) power dominates at rest.** The gripper holds position continuously and draws ~1.2 W even when the arm is stationary. The rest-period tool energy is currently distributed evenly across cycles — you may want to apportion it by cycle duration instead.

4. **Base power is not included here.** See `energy_log_base_<ts>.csv` for mobile base energy. For full-system LCA, merge base and arm files on `timestamp`.

5. **Right arm was stationary** during this recording (arm energy ≈ 0 J, tool power ~1.2 W). Include its tool energy in the system total if the right arm's gripper was powered throughout.
